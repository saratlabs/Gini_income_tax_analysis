```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Define countries and years
countries = ['Germany', 'Netherlands', 'Sweden', 'India']
years = np.arange(2000, 2024)

# Define your paths
project_path = "/content/drive/MyDrive/gini_income_tax_analysis"
data_path = f"{project_path}/data"

# -----------------------
# Simulate Gini Data
# -----------------------
gini_data = []

for country in countries:
    for year in years:
        if country in ['Germany', 'Netherlands', 'Sweden']:
            gini_pre_tax = np.random.uniform(0.45, 0.55)
            gini_post_tax = gini_pre_tax - np.random.uniform(0.10, 0.15)
        else:  # India
            gini_pre_tax = np.random.uniform(0.50, 0.55)
            gini_post_tax = gini_pre_tax - np.random.uniform(0.03, 0.07)

        gini_data.append([country, year, round(gini_pre_tax, 3), round(gini_post_tax, 3)])

gini_df = pd.DataFrame(gini_data, columns=['Country', 'Year', 'Gini_Pre_Tax', 'Gini_Post_Tax'])
gini_df.to_csv(f"{data_path}/simulated_gini_data.csv", index=False)
print("Simulated Gini data saved.")

# -----------------------
# Simulate Tax Rate Data
# -----------------------
tax_data = []

for country in countries:
    for year in years:
        if country == 'Germany':
            tax_rate = np.random.uniform(42, 46)
        elif country == 'Netherlands':
            tax_rate = np.random.uniform(38, 42)
        elif country == 'Sweden':
            tax_rate = np.random.uniform(50, 55)
        elif country == 'India':
            tax_rate = np.random.uniform(30, 35)

        tax_data.append([country, year, round(tax_rate, 2)])

tax_df = pd.DataFrame(tax_data, columns=['Country', 'Year', 'Tax_Rate'])
tax_df.to_csv(f"{data_path}/simulated_tax_rates.csv", index=False)
print("Simulated Tax Rate data saved.")

# -----------------------
# Merge and Save Merged Dataset
# -----------------------
merged_df = pd.merge(gini_df, tax_df, on=['Country', 'Year'])
merged_df.to_csv(f"{data_path}/merged_data.csv", index=False)
print("Merged dataset saved.")
```

With this we have simulated the data and have also merged the data which is a part of data preparation process which is essential for EDA (Exploratory Data Analysis) process which will be continuing further.
